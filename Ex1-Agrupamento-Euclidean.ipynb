```python
from sklearn.datasets import load_wine

wine = load_wine()

X, y = wine.data, wine.target
results = {}
RAND_results = {}
```

```python
import numpy as np
import pandas as pd
import scipy
from scipy.spatial import distance_matrix 
from scipy.spatial.distance import pdist, squareform
from scipy.cluster import hierarchy 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.metrics.cluster import adjusted_rand_score
#
from scipy.cluster.hierarchy import single,complete,average, fcluster
from sklearn.cluster import AgglomerativeClustering
```

# Euclidean

```python
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.pdist.html#scipy.spatial.distance.pdist
#https://stackoverflow.com/questions/13079563/how-does-condensed-distance-matrix-work-pdist
MatC = pdist(X, 'euclidean')
MatF = squareform(MatC)
MatF
```

# City Block

```python
Dist = pdist(X, 'cityblock')
Dist = squareform(Dist)
Dist
```

```python
agg_sing = AgglomerativeClustering(linkage ='single', n_clusters=3)
agg_sing.fit(X)

df_sl = pd.DataFrame(X, columns = ['att_1','att_2','att_3', 'att_4','att_5','att_6','att_7','att_8','att_9','att_10','att_11','att_12','att_13'])
df_sl['cluster'] = agg_sing.labels_
df_sl['silhueta'] = silhouette_samples(X, agg_sing.labels_)
df_sl
```

```python
results["SL"]=df_sl.silhueta.mean()
```

```python
C = agg_sing.labels_  # resultado do agrupamento
R = y  # organização de referência

RAND = adjusted_rand_score(C,R)
RAND_results["SL"]=RAND
```

```python
agg_comp = AgglomerativeClustering(linkage ='complete', n_clusters=3)
agg_comp.fit(X)

df_cl = pd.DataFrame(X, columns = ['att_1','att_2','att_3', 'att_4','att_5','att_6','att_7','att_8','att_9','att_10','att_11','att_12','att_13'])
df_cl['cluster'] = agg_comp.labels_
df_cl['silhueta'] = silhouette_samples(X, agg_comp.labels_)
df_cl
```

```python
results["CL"]=df_cl.silhueta.mean()
```

```python
C = agg_comp.labels_  # resultado do agrupamento
R = y  # organização de referência

RAND = adjusted_rand_score(C,R)
RAND_results["CL"]=RAND
```

# k-means

```python
#Silhueta

kmeans = KMeans(n_clusters=3,n_init=10,init='random',max_iter=300,random_state=0)
kmeans.fit(X)

df_km_s = pd.DataFrame(X, columns = ['att_1','att_2','att_3', 'att_4','att_5','att_6','att_7','att_8','att_9','att_10','att_11','att_12','att_13'])
df_km_s['cluster'] = kmeans.labels_
df_km_s['silhueta'] = silhouette_samples(X, kmeans.labels_)
df_km_s
```

```python
results["K_means"]=df_km_s.silhueta.mean()
```

```python
C = kmeans.labels_  # resultado do agrupamento
R = y  # organização de referência

RAND = adjusted_rand_score(C,R)
RAND_results["K_means"]=RAND
```

```python
melhor_result = max(results, key=results.get)
print("Melhor resultado foi: ", melhor_result,results[melhor_result])
print("Seu RAND foi: ", RAND_results[melhor_result])
```
